```python
def cs_service_bot():
    print("Hello! Welcome to the 404 Company's Service Portal. Are you a new or existing customer?")
    print("[1] New Customer") 
    print("[2] Existing Customer")
    response = input()
    if response == "1":
        new_customer()
    elif response == "2":
        existing_customer()
    else:
        print("Sorry, we didn't understand your selection")
        cs_service_bot()

def new_customer():
    print("Welcome, human!")
    print("[1] Sign up for service") 
    print("[2] Schedule a home visit") 
    print("[3] Speak to another human")
    response = input() 
    if response == "1":
        sign_up()
    elif response == "2":
        home_visit("none")
    elif response == "3":
        live_rep("sales")
    else:
        print("Sorry, we didn't understand your selection")
        new_customer()

        
def existing_customer():
    print("What kind of support do you need?")
    print("[1] Television Support")
    print("[2] Internet Support") 
    print("[3] Speak to a human")
    response = input()
    if response == "1":
        television_support()
    elif response == "2":
        internet_support()
    elif response == "3":
        live_rep()
    else:
        print("Sorry, we didn't understand your selection.")
        existing_customer()
        
        
def television_support():
    print("Select your Television problem:")
    print("[1] I can't access certain channels.")
    print("[2] My picture is blurry.")
    print("[3] I keep losing service.")
    response = input()
    if response == "1":
        print("Please check the channel lists at DefinitelyNotSinister.com. If the channel you cannot access is there, please contact a live representative.")
    elif response == "2":
        print("Try adjusting the antenna above your television set.")
    elif response == "3":
        print("Is it raining outside? If so, wait until it is not raining and then try again.")
    else:
        print("Sorry, we didn't understand your selection")
        television_support()
        
        
def internet_support():
    print("Select your Internet problem:")
    print("[1] I can't connect to the internet.")
    print("[2] My connection is very slow.")
    print("[3] I can't access certain sites.")
    response = input()
    if response == "1":
        print("Unplug your router, count to 1,000,000, and plug it back in, then give it a good whack.")
        did_that_help()
    elif response == "2":
        print("Have you considered upgrading your plan? If you give us more money we might consider a possible increase in bandwidth eventually in maybe your area.")
        did_that_help()
    elif response == "3":
        print("Try installing our VPN. Rates begin at $99/mo.")
        did_that_help()
    else:
        print("Sorry, we didn't understand your selection")
        internet_support()

        
def did_that_help():
    print("Did that help solve your issue?")
    print("Enter 'Yes' or 'No'")
    response = input()
    if response == "Yes":
        print("Thank you for using this service, have a nice day")
    elif response == "No":
        print("What would you like to try next?")
        print("[1] Speak to a human")
        print("[2] Schedule a home visit")
        response2 = input()
        if response2 == "1":
            live_rep("support")
        elif response2 == "2":
            home_visit("support")
        else:
            print("Sorry, we didn't understand your selection")
            did_that_help()
    else: 
        print("Sorry, we didn't understand your selection")
        did_that_help()
        
        
def sign_up():
    print("Great choice, human! We're excited to have you join the DNS family! Please select the package you are interested in signing up for.")
    print("[1] Bundle Deal (Internet + Cable)")
    print("[2] Internet")
    print("[3] Cable")
    response = input()
    if response == "1":
        print("You've selected the Bundle Package! Please schedule a home visit and our technician will come and set up your new service.")
        home_visit("install your cable internet bundle.")
    elif response == "2":
        print("You've selected the Internet Only Package! Please schedule a home visit and our technician will come and set up your new service.")
        home_visit("install your internet.")
    elif response == "3":
        print("You've selected the Cable Only Package! Please schedule a home visit and our technician will come and set up your new service.")
        home_visit("install your cable.")
    else:
        print("Sorry, we didn't understand your message")
        sign_up()
        
        
def home_visit(purpose):
    if purpose == "none":
        print("What is the purpose of your requested home visit?")
        print("[1] New service installation.")
        print("[2] Existing service repair.")
        print("[3] Location scouting for unserviced region.")
        response = input()
        if response == "1":
            home_visit("new install")
        elif response == "2":
            home_visit("support")
        elif response == "3":
            home_visit("scout")
    else:
        print("Please enter a date below when you are available for a technician to come to your home and " + purpose)
        response = input()
        print("Wonderful! A technician will come visit you on " + response + ". Please be available between the hours of 1:00 am and 11:00 pm.")

        
def live_rep(purpose):
    if purpose == "sales":
        print("Please hold while we connect you with a live sales representative. The wait time will be between two minutes and six hours. We thank you for your patience.")
    elif purpose == "support":
        print("Please hold while we connect you with a live support representative. The wait time will be between two minutes and six hours. We thank you for your patience.")
```
