```python
# In this project I use matplotlib to visualize simulated data for a new and hip startup
# that sells limes on the internet.

from matplotlib import pyplot as plt

# define the time series
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
x_values = range(len(months))

# numbers of limes of different species sold each month
key_limes_per_month = [92.0, 109.0, 124.0, 70.0, 101.0, 79.0, 106.0, 101.0, 103.0, 90.0, 102.0, 106.0]
persian_limes_per_month = [67.0, 51.0, 57.0, 54.0, 83.0, 90.0, 52.0, 63.0, 51.0, 44.0, 64.0, 78.0]
blood_limes_per_month = [75.0, 75.0, 76.0, 71.0, 74.0, 77.0, 69.0, 80.0, 63.0, 69.0, 73.0, 82.0]
# website visits by month
visits_per_month = [9695, 7909, 10831, 12942, 12495, 16794, 14161, 12762, 12777, 12439, 10309, 8724]


# create the figure
plt.figure(figsize=(15, 8))



# subplot 1
# website visits by month
ax1 = plt.subplot(1,2,1)
ax1.plot(x_values, visits_per_month, marker='o')
plt.xlabel('Month')
plt.ylabel('Site Visits')
ax1.set_xticks(x_values)
ax1.set_xticklabels(months)
plt.title('Site Visits per Month')


# subplot 2
# number and specie of lime sold by month
ax2 = plt.subplot(1,2,2)
ax2.plot(x_values, key_limes_per_month)
ax2.plot(x_values, persian_limes_per_month)
ax2.plot(x_values, blood_limes_per_month)
legend = ['Key', 'Persian', 'Blood']
ax2.legend(legend)
ax2.set_xticks(x_values)
ax2.set_xticklabels(months)
plt.title("Lime Sales per Month")

# adjust the plot so it looks nice
plt.subplots_adjust(wspace=0.25)
# save the plot onto your computer
#plt.savefig('lime_graphs.png')

# show the plot
plt.show()
```
